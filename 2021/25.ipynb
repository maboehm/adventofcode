```python
from collections import defaultdict

ex = """...>>>>>..."""
def parse(inp):
    return [list(l.strip()) for l in inp.strip().splitlines()]

parse(ex)
```

```python
from itertools import product
g = parse(ex)

def get(g,i,j):
    if i >= len(g):
        i -= len(g)
    if j >= len(g[0]):
        j -= len(g[0])
    return g[i][j]

def put(g,i,j, s):
    if i >= len(g):
        i -= len(g)
    if j >= len(g[0]):
        j -= len(g[0])
    g[i][j] = s

def step_right(g):
    updated = False
    i = 0
    while i < len(g):
        j = 0
        while j < len(g[0]):
            if g[i][j] == '>':
                if get(g,i,j+1) == '.':
                    g[i][j] = '.'
                    put(g,i,j+1, '>')
                    j += 1
                    updated = True
            j += 1
        i += 1
    return updated
    
def step_down(g):
    updated = False
    j = 0
    while j < len(g[0]):
        i = 0
        while i < len(g):
            if g[i][j] == 'v':
                if get(g,i+1,j) == '.':
                    g[i][j] = '.'
                    put(g,i+1,j, 'v')
                    i += 1
                    updated = True
            i += 1
        j += 1
    return updated

def step(g):
    u1 = step_right(g)
    u2 = step_down(g)
    return u1 or u2
```

```python
ex2 = """...>...
.......
......>
v.....>
......>
.......
..vvv.."""

g = parse(ex2)
for _ in range(5):
    show(g)
    step(g)
```

```python
def parse(inp):
    crabs = {}
    max_i = 0; max_j = 0
    for i, l in enumerate(inp.strip().splitlines()):
        max_i = max(i, max_i)
        for j, c in enumerate(l.strip()):
            max_j = max(j, max_j)
            if c != '.':
                crabs[(i,j)] = c
    return crabs, (max_i+1, max_j+1)

parse(ex4)
```

```python
def val_point(p, maxes):
    p0, p1 = p
    if p0 >= maxes[0]:
        p0 -= maxes[0]
    if p1 >= maxes[1]:
        p1 -= maxes[1]
    return (p0, p1)
        
def step(crabs: dict[tuple, str], maxes):
    updated = False
    new_crabs1 = {}
    for pos,c  in crabs.items():
        if c == 'v':
            new_crabs1[pos] = c
            continue
        n = val_point((pos[0], pos[1]+1), maxes)
        if n in crabs:
            new_crabs1[pos] = c
        else:
            new_crabs1[n] = c
            updated = True
    new_crabs2 = {}
    for pos,c  in new_crabs1.items():
        if c == '>':
            new_crabs2[pos] = c
            continue
        n = val_point((pos[0]+1, pos[1]), maxes)
        if n in new_crabs1:
            new_crabs2[pos] = c
        else:
            updated = True
            new_crabs2[n] = c
    return updated, new_crabs2

crabs, maxes = parse(ex4)
step(crabs, maxes)
```

```python
def show(crabs: dict, maxes):
    grid = [list("."*(maxes[1])) for _ in range(maxes[0])]
    for pos, c in crabs.items():
        grid[pos[0]][pos[1]] = c
    print("---")
    print("\n".join(["".join(l) for l in grid]))
crabs, maxes = parse(ex4)
show(crabs, maxes)
_, crabs = step(crabs, maxes)
show(crabs, maxes)
val_point((4,0), maxes)
        
```
